A <b>Decision Tree</b> is a type of supervised machine learning used to categorize or make predictions based on how a previous set of questions were answered. The model is a form of supervised learning, meaning that the model is trained and tested on a set of data that contains the desired categorization.

```python
iris = datasets.load_iris()
```

```python
data=pd.DataFrame(iris['data'],columns=["petal_length","petal_width","sepal_length","sepal_width"])
data['species']=iris['target']
data.head()
```
