<b>Exploratory Data Analysis (EDA)</b> is a crucial step in the data analysis process. It involves examining and visualizing data sets to gain a better understanding of their main characteristics, patterns, and relationships.EDA helps data analysts and scientists to uncover initial insights, identify anomalies, and formulate hypotheses for further investigation.  
    <p>Overall, Exploratory Data Analysis provides a comprehensive and systematic approach to understanding and analyzing data before proceeding to more advanced modeling or inferential techniques. It helps analysts gain valuable insights, validate assumptions, and make informed decisions based on data-driven evidence.<p/>

### Step 1 - Import Necessary libraries

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Step 2 - Load Dataset to Python Environment

```python
data = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin1')
data.head()
```

```python
# Renaming necessary columns for convenience of processing data
data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                          'region_txt':'Region','attacktype1_txt':'Attack_Type','target1':'Target','nkill':'Killed',
                          'nwound':'Wounded','gname':'Group','targtype1_txt':'Target_Type',
                          'weaptype1_txt':'Weapon_type','latitude':'Latitude',
                          'longitude':'Longitude','target1':'Target','city':'City'},inplace=True)
```

```python
# percentage of missing values in our dataset
missing_values = (((data.isnull().sum()).sum())/data.size)*100
missing_values
```

More than 50% values are Null, we need to clean the Dataset first

```python
# creating dataframe with necessary columns only
data = data[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','Attack_Type','Killed',
              'Wounded','Group','Target','Target_Type','Weapon_type']]
data.head(10)
```

```python
# Finding the number of unique values in each necessary columns
for i in data.columns:
    print(i,data[i].nunique())
```

```python
data.info()
```

As we can see, 'Wounded' and 'Killed' Attributes have lots of Null values, we need to fill null values

```python
# Checking the distribution of columns
freqgraph = data.select_dtypes(include = ['float','int'])
freqgraph
```

```python
# Plotting the frequency graph
freqgraph.hist(figsize=(20,15))
plt.show()
```

```python
# Filling the missing values with median based on their frequency distribution
data['Wounded'] = data['Wounded'].fillna(data['Wounded'].median()).astype(int)
data['Killed'] = data['Killed'].fillna(data['Killed'].median()).astype(int)
```

```python
data.info()
```

### Step 4 - Exploratory Data Analysis

### 4.1. Univariate Analysis

```python
data['Attack_Type'].value_counts()
```

```python
(data['Attack_Type'].value_counts()/data.shape[0])*100
```

```python
# 10 most attacked targets

plt.figure(figsize = (8,4))
sns.histplot(data['Attack_Type'], palette='cubehelix')
plt.title('Attack Types',fontsize=15)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

<b>Observation</b>  
1. 50% of the Attack happened with Bombing/Explosion  
2. 10-20% of the Attack happened with Armed Assault and Assassination rest are in between 0-6%

### Terrorist Targets: The Ten Countries Which Suffer Most From Terrorism

```python
data.Country.value_counts()[:10]
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x = data['Country'].value_counts()[:10].index,y = data['Country'].value_counts()[:10].values,palette='viridis')
plt.title('Top 10 Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

According to the graph, Iraq is the most affected country by terrorism, then comes Pakistan and Afganistan.

```python
(data['Target_Type'].value_counts().head(10)/data['Target_Type'].shape[0])*100
```

```python
# 10 most attacked targets
plt.figure(figsize = (8,4))
sns.barplot(x = data['Target_Type'].value_counts().head(10).index, y = data['Target_Type'].value_counts().head(10).values, 
            palette='magma')
plt.title('Top 10 most attacked targets',fontsize=15)
plt.xlabel('Targets',fontsize=15)
plt.ylabel('Number of attacks',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

<b>Observation</b>  
1. The most Attacked Target is 'Private Citizens & Property' which is approximately 25%  
2. 10-20% is the Target which is 'Military','Police', 'Government', 'Business'

```python
# regions with most attacks 
plt.figure(figsize = (8,4))
sns.barplot(x = data['Region'].value_counts().index, y = data['Region'].value_counts().values,palette='flare')
plt.title('Most Attacked Regions',fontsize=15)
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
# States with most attacks 
plt.figure(figsize = (8,4))
sns.barplot(x = data['State'].value_counts().head(15).index, y = data['State'].value_counts().head(15).values,palette='viridis')
plt.title('Top 10 Most Attacked States',fontsize=15)
plt.xlabel('States',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15, 5)
```

<b>Observation</b>  
1. Most Attacked Region is 'Middle East and North Africa' then comes South Asia  
2. Most Attacked State is 'Baghdad'

```python
# % of different weapon types used
(data['Weapon_type'].value_counts().head()/data['Weapon_type'].shape[0])*100
```

```python
# Top 5 most used weapons types in terror attacks
plt.figure(figsize=(8,4))
sns.barplot(x = data['Weapon_type'].value_counts().head().index,y = data['Weapon_type'].value_counts().head().values,palette='flare')
plt.title('Top 5 Most Used Weapon types',fontsize=15)
plt.xlabel('Weapon Types',fontsize=15)
plt.ylabel('Number of Weapon Used',fontsize=15)
plt.show()
```

Almost 50% of Weapon used is 'Explosives' whereas 32% 'Firearms' used in the terrorism

```python
# Number of attacks each year
x_year = data['Year'].unique()
y_count_years = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (12,7))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'dark:salmon_r')
plt.xticks(rotation = 90)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

<b>Observation</b>  
1. 2014 appears to be witness of a huge terrorist attacks  
2. 2014 onwards recorded a large increase in attacks of terrorism.

```python
# Number of deaths happened in different years
plt.figure(figsize = (12,7))
data.groupby(['Year'])['Killed'].sum().plot(kind='bar',colormap='summer')
plt.title('Number of Deaths in different years',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Deaths',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

Number of deaths caused by terrorism is more between 2014 and 2018, with on an average >50K people being killed. High peak went  
to 2014 with 45K+ deaths.

```python
# Terrorist Activities by Region in Each Year
plt.figure(figsize = (12,7))
sns.kdeplot(x = data['Year'], hue = data['Region'])
plt.title('Terrorist Activities by Region in each Year',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Frequency of Attacks',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

### 4.2. Bivariate Analysis

```python
plt.subplot(1,2,1)

data.groupby(['Region'])['Wounded'].sum().sort_values(ascending = False).plot(kind='bar',colormap='rocket')
plt.title('Regions having Wounded People',fontsize=15)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Number of Wounded',fontsize=15)
plt.xticks(rotation=90)

plt.subplot(1,2,2)
data.groupby(['Region'])['Killed'].sum().sort_values(ascending = False).plot(kind='bar',colormap='rocket')
plt.title('Regions having Killed People',fontsize=15)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Number of Killed',fontsize=15)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15, 5)
```

'Middle East & North Africa' has the most Killed people (>1.4 Lacs) and wounded people (>2 lacs)

```python
plt.subplot(1,2,2)

df1=data.groupby(['Country'])['Wounded'].sum().sort_values(ascending = False).head(10).plot(kind='bar',colormap='winter')
plt.title('Country having Wounded People',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Number of Wounded',fontsize=15)
plt.xticks(rotation=90)

plt.subplot(1,2,1)
data.groupby(['Country'])['Killed'].sum().sort_values(ascending = False).head(10).plot(kind='bar',colormap='winter')
plt.title('Country having Killed People',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Number of Killed',fontsize=15)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15, 5)
```

'Iraq' has the most Killed people (>70K) and wounded people (>1.2 lacs) Afganistan, pakistan and India all saw peak deaths.  


```python
# Top 10 Terrorist Organizations with Highest Terror Attacks
plt.figure(figsize=(8,8))
sns.barplot(x = data['Group'].value_counts()[1:11].values, y = data['Group'].value_counts()[1:11].index,palette='rocket')
plt.title('Top 10 Terrorist Organization with Highest Terror Attacks',fontsize=15)
plt.xlabel('Number of Attacks',fontsize=15)
plt.ylabel('Terrorist Groups',fontsize=15)
plt.show()
```

```python
# Creating new dataframe for Taliban
data_tal = data[data.Group == 'Taliban']
data_tal
```

```python
# Terror Attacks by Taliban over the Years
plt.figure(figsize = (10,5))
sns.barplot(x = data_tal['Year'].value_counts().index,y = data_tal['Year'].value_counts().values, palette = 'viridis')
plt.title('Terror Attack over the years by Taliban',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
data_tal.groupby(['Year'])['Killed'].sum().plot(kind='bar',colormap='RdBu')
plt.title('People Killed by Taliban over the Years',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of people killed',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

Taliban killed more than 5000 people during 2015 after that a decreasing trend could be seen but still the numbers are very high
